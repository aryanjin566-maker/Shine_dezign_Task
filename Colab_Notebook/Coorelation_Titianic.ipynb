import joblib
import numpy as np

# Load model and scaler
model = joblib.load("/content/diabetes_model.pkl")
scaler = joblib.load("/content/scaler.pkl")

# Input values (same order as training)
Pregnancies = int(input("Pregnancies: "))
Glucose = float(input("Glucose: "))
BMI = float(input("BMI: "))
Age = int(input("Age: "))
Insulin = float(input("Insulin: "))

# Create input array
data = np.array([[Glucose, BMI, Age, Pregnancies, Insulin]])

# Scale input
data_scaled = scaler.transform(data)

# Predict
prediction = model.predict(data_scaled)

if prediction[0] == 1:
    print("⚠️ Patient likely has Diabetes")
else:
    print("✅ Patient likely does NOT have Diabetes")
