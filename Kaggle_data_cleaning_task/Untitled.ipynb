```python
# Calculate percentage of missing values
missing_percentage = (zomato_df.isnull().sum() / len(zomato_df)) * 100
missing_percentage
```

```python
#Dropping Unnecessary Columns
zomato_df.drop(columns = ['url','address','phone', 'location','dish_liked','menu_item','cuisines','approx_cost(for two people)','reviews_list'],inplace = True)
print("\nRemaining Columns after Dropping:")
print(zomato_df.columns)
```

```python
zomato_df['rate'] = zomato_df['rate'].replace(['NEW', '-', 'Not Rated Yet', 'Excellent'], None)
```

```python
import re

def extract_rating(val):
    match = re.search(r'(\d+\.\d+)', str(val))
    return float(match.group(1)) if match else None

zomato_df['rate'] = zomato_df['rate'].apply(extract_rating)

print(zomato_df['rate'].unique())   
print(zomato_df['rate'].dtype) 
```

```python
rate_mean = zomato_df['rate'].mean()
rate_mean
```

```python
rate_median = zomato_df['rate'].median()
rate_median
```

```python
zomato_df['rate'] = zomato_df['rate'].fillna(rate_mean)
zomato_df['rate'].isnull().sum()
```

```python
zomato_df['rest_type'].unique()
```

```python
zomato_df['rest_type'] = zomato_df['rest_type'].fillna('Others')
```

```python
zomato_df.isnull().sum()
```

```python
zomato_df['online_order'] = zomato_df['online_order'].replace({'Yes': 1, 'No': 0})
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
zomato_df['book_table'] = le.fit_transform(zomato_df['book_table'])
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()

encoded = ohe.fit_transform(zomato_df[['location']])

encoded_df = pd.DataFrame(
    encoded,
    columns=ohe.get_feature_names_out(['location'])
)

zomato_df = pd.concat([zomato_df, encoded_df], axis=1)
zomato_df.drop('location', axis=1, inplace=True)
```
